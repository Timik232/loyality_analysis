```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
missing_values = data_accounts.isnull().sum()
total_values = len(data_accounts)
missing_percent = (data_accounts.isnull().sum() / len(data_accounts)) * 100
plt.figure(figsize=(10, 12))
sns.barplot(x=missing_percent.index, y=missing_percent.values, hue=missing_percent.index, dodge=False)
plt.title('Процент отсутствующих данных по столбцам')
plt.xlabel('Столбцы')
plt.ylabel('Процент пропущенных значений')
plt.xticks(rotation=45)
for i in range(len(missing_percent)):
    plt.text(i, missing_percent.iloc[i], f'Отсутствующих: {missing_values.iloc[i]}\nВсего: {total_values}', ha='center', va='bottom')
plt.ylim(0, missing_percent.max() * 1.1)
plt.savefig("missing_pnsn.png",)
plt.show()
```

```python
missing_values = data.isnull().sum()
total_values = len(data)

# Вычисляем процент пропущенных значений
missing_percent = (missing_values / len(data)) * 100

# Отфильтровываем столбцы, которые содержат пропущенные значения
missing_data = missing_percent[missing_percent > 0]

plt.figure(figsize=(10, 14))
sns.barplot(x=missing_data.index, y=missing_data.values, hue=missing_data.index, dodge=False)
plt.title('Процент отсутствующих данных по столбцам')
plt.xlabel('Столбцы')
plt.ylabel('Процент пропущенных значений')
plt.xticks(rotation=45)
for i in range(len(missing_data)):
    plt.text(i, missing_data.iloc[i], f'Отсутствующих: {missing_values[missing_data.index[i]]}\nВсего: {total_values}', ha='center', va='bottom')
plt.ylim(0, missing_data.max() * 1.1)
plt.savefig("missing_values.png")
plt.show()
```

```python
#drop("client_id").drop("npo_account_id")
categorical_columns = data.select_dtypes(include=['object']).columns
df = data.drop(columns=categorical_columns)
# npo и ops ни с чем не кореллирует, это приёмник счёта
# email, телефон,адрес, почтовый код и итоговая сумма также выкидывается
columns = ["email", "phone_number", "postal_code", "fact_addrss", "churn", "assignee_ops", "assignee_npo"]
df = df.drop(columns=columns)
corr_matrix = df.corr()
plt.figure(figsize=(10, 14))
sns.heatmap(corr_matrix, cmap='coolwarm', fmt=".2f", linewidths=.05)
plt.title('Матрица корреляций')
plt.savefig("corellations.png")
plt.show()
```

```python
categorical_columns = data.select_dtypes(include=['object']).columns
id_columns = ["client_id", "npo_account_id"]
df = data[categorical_columns].drop(columns=id_columns)

for column in df.columns:
    # Ограничиваем количество категорий для отображения
    top_categories = df[column].value_counts().index[:10]
    df_limited = df[df[column].isin(top_categories)]
    
    # Вычисляем процентное соотношение каждой категории
    category_counts = df_limited[column].value_counts()
    category_percentage = (category_counts / category_counts.sum()) * 100
    
    plt.figure(figsize=(10, 8))
    ax = sns.countplot(y=column, data=df_limited, order=df_limited[column].value_counts().index)
    plt.title(f'Гистограмма для {column}')
    plt.xlabel('Количество')
    plt.ylabel(column)
    
    # Добавляем числовые значения на вершину каждого столбца
    for i, v in enumerate(category_counts):
        ax.text(v, i, f' {category_percentage.iloc[i]:.1f}%', va='center')
    plt.xlim(0, category_counts.max() * 1.2)
    plt.savefig(f"{column}.png")
    plt.show()
```

```python
categorical_columns = data.select_dtypes(include=['object']).columns
id_columns = ["client_id", "npo_account_id"]
df_frst_pmnt_date = data[categorical_columns].drop(columns=id_columns)
df_lst_pmnt_date_per_qrtr = df_frst_pmnt_date.copy()

# Преобразуем столбец даты в формат datetime
df_frst_pmnt_date['date'] = pd.to_datetime(df_frst_pmnt_date['frst_pmnt_date'])
df_lst_pmnt_date_per_qrtr['date'] = pd.to_datetime(df_lst_pmnt_date_per_qrtr['lst_pmnt_date_per_qrtr'])

print(df_frst_pmnt_date['date'])
```

```python
df_frst_pmnt_date['year'] = df_frst_pmnt_date['date'].dt.year
df_frst_pmnt_date['month'] = df_frst_pmnt_date['date'].dt.month
month_names = {1: 'Январь', 2: 'Февраль', 3: 'Март', 4: 'Апрель', 5: 'Май', 6: 'Июнь', 7: 'Июль', 8: 'Август', 9: 'Сентябрь', 10: 'Октябрь', 11: 'Ноябрь', 12: 'Декабрь'}
df_frst_pmnt_date['month'] = df_frst_pmnt_date['month'].map(month_names)
df_frst_pmnt_date['year'] = df_frst_pmnt_date['year'].astype(str).str.rstrip('.0')
# df_frst_pmnt_date['month'] = df_frst_pmnt_date['month'].astype(str).str.rstrip('.0')
print(df_frst_pmnt_date['year'])
print(df_frst_pmnt_date['month'])
```

```python
# Извлекаем год и месяц
df_lst_pmnt_date_per_qrtr['year'] = df_lst_pmnt_date_per_qrtr['date'].dt.year
df_lst_pmnt_date_per_qrtr['month'] = df_lst_pmnt_date_per_qrtr['date'].dt.month
month_names = {1: 'Январь', 2: 'Февраль', 3: 'Март', 4: 'Апрель', 5: 'Май', 6: 'Июнь', 7: 'Июль', 8: 'Август', 9: 'Сентябрь', 10: 'Октябрь', 11: 'Ноябрь', 12: 'Декабрь'}
df_lst_pmnt_date_per_qrtr['month'] = df_lst_pmnt_date_per_qrtr['month'].map(month_names)
df_lst_pmnt_date_per_qrtr['year'] = df_lst_pmnt_date_per_qrtr['year'].astype(str).str.rstrip('.0')
# df_lst_pmnt_date_per_qrtr['month'] = df_lst_pmnt_date_per_qrtr['month'].astype(str).str.rstrip('.0')
print(df_lst_pmnt_date_per_qrtr['year'])
print(df_lst_pmnt_date_per_qrtr['month'])
```

```python

# Теперь вы можете получить статистику по годам и месяцам, как вы делали ранее
for column in ['year', 'month']:
    # Ограничиваем количество категорий для отображения
    top_categories = df_frst_pmnt_date[column].value_counts().index[:10]
    df_frst_pmnt_date_limited = df_frst_pmnt_date[df_frst_pmnt_date[column].isin(top_categories)]
    
    # Вычисляем процентное соотношение каждой категории
    category_counts = df_frst_pmnt_date_limited[column].value_counts()
    category_percentage = (category_counts / category_counts.sum()) * 100
    
    plt.figure(figsize=(10, 8))
    ax = sns.countplot(y=column, data=df_frst_pmnt_date_limited, order=df_frst_pmnt_date_limited[column].value_counts().index)
    plt.title(f'Гистограмма для {column} по frst_pmnt_date')
    plt.xlabel('Количество')
    plt.ylabel(column)
    
    # Добавляем числовые значения на вершину каждого столбца
    for i, v in enumerate(category_counts):
        ax.text(v, i, f' {category_percentage.iloc[i]:.1f}%', va='center')
    plt.xlim(0, category_counts.max() * 1.2)
    plt.savefig(f"{column}_frst_pmnt_date.png")
    plt.show()
```
